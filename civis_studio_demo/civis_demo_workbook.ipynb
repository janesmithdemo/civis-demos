## Pull and View Data from public.iris Table
This example uses the Civis Python API to query the `public.iris` table and display the first few rows using pandas.

```python
import civis
import pandas as pd

# Replace with your database ID if different
DATABASE_ID = 1234  # Example: 1234, update as needed

# SQL to query the iris table
sql = "SELECT * FROM public.iris LIMIT 10"

# Run the query using Civis API
result = civis.io.read_civis(sql, DATABASE_ID, use_pandas=True)

# Display the first few rows
result.head()
```
